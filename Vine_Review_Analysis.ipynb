```python
import pandas as pd
```

```python
# Read in the data
df = pd.read_csv('vine_table.csv')
df.head()
```

```python
# Filter DataFrame where the total_votes count is equal to or greater than 20
df1 = df.loc[df["total_votes"] >= 20]
df1.head(5)
```

```python
# Filter DataFrame where the number of helpful_votes divided by total_votes is equal to or greater than 50%.
df2 = df1.loc[df1["helpful_votes"] / df1["total_votes"] >= 0.5]
df2.head(5)
```

```python
# Filter DataFrame where a review was written as part of the vine program
df3 = df2.loc[df2["vine"] == 'Y']
df3.head(5)
```

```python
# Filter DataFrame where a review was not written as part of the vine program
df4 = df2.loc[df2["vine"] == 'N']
df4.head(5)
```

```python
# Determine the total number of reviews
totalcount = df.review_id.count()
totalcount
```

```python
# Determine the number of 5-star reviews
fivestar = df.loc[(df['star_rating']== 5)]["star_rating"].count()
fivestar
```

```python
# Determine the percentage of 5-star reviews for the two types of review (paid vs unpaid).
paidtotal = df3.review_id.count()
paidfivestar = df3.loc[(df3['star_rating']== 5)]["star_rating"].count()
paidpercentage = paidfivestar/paidtotal

unpaidtotal = df4.review_id.count()
unpaidfivestar = df4.loc[(df4['star_rating']== 5)]["star_rating"].count()
unpaidpercentage = unpaidfivestar/unpaidtotal

paidpercentage*100
```
